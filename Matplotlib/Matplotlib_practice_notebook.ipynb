```python
data =  pd.read_csv('topic_2_data.csv')
print(data.columns)
data.shape
```

```python
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']
```

```python
# Counter() is python module/method that keeps the count of number of occurences of one or many items
language_counter = Counter()
```

```python
# Counter.update() is method to update the count
for response in lang_responses:
    language_counter.update(response.split(';'))
    
print(language_counter)
```

```python
languages, popularity = [], []
for items in language_counter.most_common(20):
    languages.append(items[0])
    popularity.append(items[1])
    
# below is executed to set the data in ascending order
languages.reverse()
popularity.reverse()

print(languages, popularity)
```

```python
plt.barh(languages, popularity)

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()

plt.show()
```

# 5. Filling areas on Line plots


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('topic_5_data.csv')

ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

# below code will shade the area below the graph plot
plt.fill_between(ages, py_salaries, alpha=0.25) # aplha attribute represents the transparency in the plot

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()

plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
# the area shaded above the graph plot represent that the graph is below the overall_median
# the area shaded below the graph means that the graph plot is above the overall_median
plt.fill_between(ages, py_salaries, overall_median, alpha=0.25) 

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
# below code will give different color scheme to the shaded area
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, 
                 alpha=0.25, label='Above average')
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries <= overall_median), interpolate=True, 
                 color='red', alpha=0.25, label='Below average')

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()

plt.show()
```

```python
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

#below code will the shade the area based on py_salaries and dev_salaries
plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries > dev_salaries), interpolate=True, 
                 alpha=0.25, label='Above average')
plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries <= dev_salaries), interpolate=True, 
                 color='red', alpha=0.25, label='Below average')

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()

plt.show()
```

# 8. Plotting Time Series Data

```python
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mlp_dates
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
# linestyle='solid' connects the plot or create a line plot
# marker='None' will remove the circles on the plot
plt.plot_date(dates, y, linestyle='solid')

# different alignment for x-axis label
plt.gcf().autofmt_xdate()

plt.title('Time-Series Plot')
plt.xlabel('Dates')
plt.ylabel('Y-axis')

plt.tight_layout()

plt.show()
```

```python
plt.plot_date(dates, y, linestyle='solid')

# different alignment for x-axis label
plt.gcf().autofmt_xdate()

# different format of date for x-axis label
date_format = mlp_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)

plt.title('Time-Series Plot')
plt.xlabel('Dates')
plt.ylabel('Y-axis')

plt.tight_layout()

plt.show()
```

```python
data = pd.read_csv('topic_8_data.csv')

# to change the type of Date column from string to datetime
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']
```

```python
plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```

# 9. Plotting live data in real-time

```python
import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
x_vals = []
y_vals = []

index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    
    plt.cla() # this line of code will clear the plot before plotting new graph
    plt.plot(x_vals, y_vals)
```

```python
ani = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.tight_layout()

plt.show()
```

# 10. Subplots 

### Figure are the containers holding our plots and Axes are the actual plot

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('topic_5_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
fig, ax = plt.subplots()

ax.plot(ages, py_salaries, color='#5a7d9a', linewidth=3, label='Python')
ax.plot(ages, js_salaries, color='#adad3b', linewidth=3, label='JavaScript')

ax.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

ax.legend()

ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, color='#5a7d9a', linewidth=3, label='Python')
ax2.plot(ages, js_salaries, color='#adad3b', linewidth=3, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, color='#5a7d9a', linewidth=3, label='Python')
ax2.plot(ages, js_salaries, color='#adad3b', linewidth=3, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
